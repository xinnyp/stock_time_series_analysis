```python
import pandas as pd
import plotly.plotly as py
import cufflinks as cf

cf.go_offline()

df = pd.read_csv('AMD_stock_prices.csv', index_col='date')
df.head()
```

```python
df.close.iplot(kind='scatter')
```

```python
df.iplot(kind='candlestick', open='open', high='high', low='low',close='close')
```

```python
nvdia = pd.read_csv('NVDIA_stock_prices.csv', index_col = 'date')
nvdia.head()
```

```python
intel = pd.read_csv('intel_stock_prices.csv', index_col = 'date')
intel.head()
```

```python
stocks = pd.DataFrame({'NVIDIA':nvdia.close, 'AMD':df.close, 'INTEL':intel.close})
stocks.head()
```

```python
stocks.iplot()
```

Transform the data - consider the stock's return since the beginning of the period of interest. This allows us to see how profitable each stock was since the beginning of the period.

Plots seems correlated.

```python
stock_return = stocks.apply(lambda x:x/x[0])
```

```python
stock_return.iplot(hline=1)
```

log differences have a desirable property: the sum of the log differences can be interpreted as the total change (as a percentage) over the period summed (which is not a property of the other formulations; they will overestimate growth). Log differences also more cleanly correspond to how stock prices are modeled in continuous time.

```python
import numpy as np
stock_change = stocks.apply(lambda x: np.log(x) - np.log(x.shift(1)))
stock_change.head()
```

```python
stock_change.iplot(hline=0)
```
